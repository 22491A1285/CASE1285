```python
import pandas as pd
import matplotlib.pyplot as plt
ajay=pd.read_csv('energy.csv')
print(ajay)
srh=ajay[ajay['State']=='Andhra']
rcb=ajay[ajay['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='yellow')
diag[1].plot(rcb['Date'],rcb['Consumption'])

srh=ajay[ajay['State']=='Telangana']
rcb=ajay[ajay['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='yellow')
diag[1].bar(rcb['Date'],rcb['Consumption'])

state1='Kerala'
state2='Delhi'
tdp=[state1,state2]
srh=ajay[ajay['State']=='Kerala']['Consumption'].sum()
rcb=ajay[ajay['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]
fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp,autopct='%1.1f%%')
```
