```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
vinay=pd.read_csv('energy.csv')
print(vinay)
state1=vinay[vinay['State']=='Andhra']
state2=vinay[vinay['State']=='Sikkim']
fig,diag=plt.subplots(2)
diag[0].plot(state1['Date'],state1['Consumption'])
diag[1].plot(state2['Date'],state2['Consumption'])
```
